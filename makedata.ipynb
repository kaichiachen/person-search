```python
import sys
import numpy as np
def get_img_data(img_name):
    bboxes = []
    for i in images['Img'][0]:
        if i[0][0] == img_name:
            for box in i[2]:
                for b in box:
                    bbox = list(b[0][0].astype(np.int64))
                    bbox[2], bbox[3] = bbox[0] + bbox[2], bbox[1] + bbox[3]
                    bboxes.append(bbox)
    return bboxes
                
d = data.copy()
print(d==data)
i = 0
for dd in d:
    sys.stdout.write('\r'+str(i)+'/'+str(len(d)))
    i+=1
    for dic in dd:
        bbox = get_img_data(dic['image'].split('/')[-1])
        #print(bbox)
        #print(dic['boxes'])
        #print(bbox == dic['boxes'])
        #print('-----')
        dic['boxes'] = bbox
print(d==data)
```
